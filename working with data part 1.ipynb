# Concat

```python
json_obj = {'id' : ['1','2','3','4','5'] , 
           'Feature1' : ['A' , 'C' , 'E' , 'G' , 'I' ],
           'Feature2' : ['B' , 'E' , 'F' , 'H' , 'J']
           
           }
df_1 = pd.DataFrame(json_obj , columns = ['id' , 'Feature1' , 'Feature2'])

json_obj2 = {
        'id': ['1', '2', '6', '7', '8'],
        'Feature1': ['K', 'M', 'O', 'Q', 'S'],
        'Feature2': ['L', 'N', 'P', 'R', 'T']}
df_2 = pd.DataFrame(json_obj2) # you dont have to specify the columns if you want all of them
df_2
```

```python
# how to make join for your data based on a key ,
#note how featuer 3 gonna be added with respect to id column values
df_merge_col = pd.merge(df_vertical , df_3 , on ='id')
df_merge_col
```

```python
# sometimes the key you want to merge based on has different name that the other data frame
# how to be more specific for joining column

df_merge_keys = pd.merge(df_vertical , df_3 , right_on = 'id' ,left_on = 'id')
df_merge_keys
```

```python
# here we are going to use merge to do the same job of joins in SQL
# we have three types of joins in sql and they are all the same in pandas merge
#full outer , inner , right  and left joins
```

```python
# first: full outer join
# in this type of join we are making a simple join but we are ging to include all data form both dataframes 

# you gonna notice that some of the rows has NaN values that it doest not has a mach on ther other data frame and vice versa
df_outer = pd.merge(df_1, df_2 , on= 'id', how = 'outer')
df_outer 
```

```python
# incase you want to label your data you can use the suffixes argument
df_merged_labeled_cols = pd.merge(df_1 , df_2 , on = 'id' , how = 'outer' , suffixes = ('_left' , '_right'))
df_merged_labeled_cols
```

```python
#second: inner join
df_inner = pd.merge(df_1 , df_2 , on = 'id' , how = 'inner')
df_inner # only the maching keys will generate the join otherwise you it will deleted 
```

```python
# third : right join
df_right = pd.merge(df_1 , df_2 , on = 'id' , how = 'right')
df_right # take all the right and only matching from the left 
```

```python
# third : left join
df_left = pd.merge(df_1 , df_2 , on = 'id' , how = 'left')
df_left # take all the left and only matching from the right 
```

```python
# finally what if we want to make joins based on the index of a dataframe
df_index = pd.merge(df_1, df_2 , right_index = True ,  left_index = True)
df_index
```

```python
trades = pd.DataFrame({
    'time': pd.to_datetime(['20160525 13:30:00.023',
                            '20160525 13:30:00.038',
                            '20160525 13:30:00.048',
                            '20160525 13:30:00.048',
                            '20160525 13:30:00.048']),
    'ticker': ['MSFT', 'MSFT','GOOG', 'GOOG', 'AAPL'],
    'price': [51.95, 51.95,720.77, 720.92, 98.00],
    'quantity': [75, 155,100, 100, 100]},
    columns=['time', 'ticker', 'price', 'quantity'])


quotes = pd.DataFrame({
    'time': pd.to_datetime(['20160525 13:30:00.023',
                            '20160525 13:30:00.023',
                            '20160525 13:30:00.030',
                            '20160525 13:30:00.041',
                            '20160525 13:30:00.048',
                            '20160525 13:30:00.049',
                            '20160525 13:30:00.072',
                            '20160525 13:30:00.075']),
    'ticker': ['GOOG', 'MSFT', 'MSFT','MSFT', 'GOOG', 'AAPL', 'GOOG','MSFT'],
    'bid': [720.50, 51.95, 51.97, 51.99,720.50, 97.99, 720.50, 52.01],
    'ask': [720.93, 51.96, 51.98, 52.00,720.93, 98.01, 720.88, 52.03]},
    columns=['time', 'ticker', 'bid', 'ask'])
```

```python
trades
```

```python
quotes
```

```python
# lets see how we can merge knowing that we have a time_series info in both of data frames
df_merge_asof = pd.merge_asof(trades , quotes , on = 'time' , by = 'ticker')

df_merge_asof
# thre result will be based on ticker for matching and time for 
# the nan value form aapl because there is no trads sequence bigger than the one on quotes
```

```python
# lets see how we can merge knowing that we have a time_series info in both of data frames
# adding tollerance with 2 mili  seconds if further igore .. and thats why we are expecting more nulls
df_merge_asof_tolerance = pd.merge_asof(trades , quotes , on = 'time' , by = 'ticker' , tolerance = pd.Timedelta('2ms'))

df_merge_asof_tolerance
```

# Pivoting

```python
# pivoting is just a way to convert data into columns of course this is depend on how frequent data are repeated and its more suitable for categorical daa 
```

```python
df = pd.DataFrame({'foo': ['one', 'one', 'one', 'two', 'two',
                           'two'],
                   'bar': ['A', 'B', 'C', 'A', 'B', 'C'],
                   'baz': [1, 2, 3, 4, 5, 6],
                   'zoo': ['x', 'y', 'z', 'q', 'w', 't']})
df
```

```python
pd.pivot_table(data = df , columns = 'foo' ,index = 'bar', values = 'baz' ) # by default the index is grouped and mean is calcualted 
```

```python
# to change the default you can use aggfunc arguments
pd.pivot_table(data = df , columns = 'foo' ,index = 'bar', values = 'baz'  , aggfunc = 'count')
```

```python
pd.pivot_table(data = df , columns = 'foo' ,index = 'bar', values = 'baz'  , aggfunc = 'sum')
```
