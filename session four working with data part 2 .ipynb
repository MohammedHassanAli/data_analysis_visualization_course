# 1. group by


```python
dframe = pd.DataFrame({'K1' : list('XXYYZ') , 'K2' : 'alpha,beta,alpha,beta,alpha'.split(',') , 'dataset1' : np.random.randn(5),
                      'dataset2' : np.random.randn(5)})
dframe
```

```python
group1 = dframe['dataset1'].groupby(dframe['K1'])
group1
```

```python
# lets take the mean of values of dframe dataset1 for x y and z
group1.mean()
```

```python
cities = np.array(['shg' , 'cairo' , 'cairo' , 'shg' , 'shg'])
month = np.array(['JAN' , 'FEB' , 'JAN' , 'FEB' , 'JAN'])
```

```python
# what if we used the arrays to work as grouping keys for the data set
# the important thing here is to put in mind that these arrays will be used as lables for the existing data and based on this story the data is going to be grouped
dframe['dataset1'].groupby([cities,month]).mean()
```

```python
dframe
```

```python
# how to group on k1 first and k2 then
dframe.groupby(['K1' , 'K2']).mean()
```

```python
# to see how many entites that was grouped for each label you use 
dframe.groupby(['K1' , 'K2']).size()
```

```python
# to itterate throught each graoup and there labels 
for name,group in dframe.groupby('K1'):
    print ("this is the group %s" %name)
    print(group)
    print('\n')
```

```python
# in case you need to make an itteration through a grouped data frame based on tow features 
for (name1, name2 ), group in dframe.groupby(['K1' , 'K2']):
    print ("this is the group1 %s group2 %s" %(name1 , name2) )
    print(group)
    print('\n')
```

```python
# how to split your data frame into multiple dataframes grouped 
group_dict  = dict(list(dframe.groupby('K1')))
group_dict
```

```python
group_dict['X']
```

```python
group_dict['Z']
```

# 2. group by on dict and series


```python
animals = pd.DataFrame(np.arange(16).reshape(4 , 4) , 
                       columns = ['W' , 'X' , 'Y' , 'Z'] , 
                       index = ['Dog' , 'Cat' , 'Bird' , 'Mouse'])
```

```python
animals
```

```python
animals.loc['Cat',['W' , 'Y']] = np.nan
animals
```

```python
#this how we can use dictionaries to make groupings
behavior_map = {'W' : 'good' , 'X': 'bad' , 'Y':'good' , 'Z' : 'bad'}
```

```python
animals_col = animals.groupby(behavior_map , axis = 1)
animals_col.sum()
```

```python
# how to use series to do grouping
behavior_series = pd.Series(behavior_map)
```

```python
animals.groupby(behavior_series,axis = 1).sum()
```

# 4. splitting applying and combining


```python
def ranker(df):
    df['alc_content_rank'] = np.arange(len(df)) + 1
    return df

# the privous function is gakeing a data frame and returning it with a new columns that wold the rank based on lenght of data frame
```

```python
# sort the data frame based on alchol content and then group them by quality .. apply ranker over the dataframe
df_wine.sort_values('alcohol' , ascending = False , inplace = True)
df_wine = df_wine.groupby('quality').apply(ranker)
```

```python
df_wine
```

# 5. cross tabulation

```python
# cross tabulation is nothing than a group by with count for respresinting the frequence of your data .. also its making a roll up for these groups
```

```python
df = pd.DataFrame({'sample' : np.arange(1,7) , 
                  'animal' : ['dog' , 'dog' , 'cat' , 'cat' , 'dog' , 'cat'] , 
                  'intelligence' : ['smart' , 'smart' , 'dumb' , 'dumb' , 'dumb' , 'smart']})
```

```python
df
```

```python
pd.crosstab(df.animal ,  df.intelligence , margins =True)
```

```python
pd.crosstab(df_wine.quality ,  df_wine.alc_content_rank , margins =True)
```
