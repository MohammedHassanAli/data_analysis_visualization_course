#  4. box plots and violin plots


```python
# box plot are blotes taht shos you distrubution mean and outlier on your data thats why its so powerful
# read more about it on wekibedia 
# to plot a boxplot you can use
arr1 = np.random.randn(100)
arr2 = np.random.randn(100)

sns.boxplot(data = [arr1 ,  arr2])
```

```python
# the points that are aligned with the box seperatly represents the outliers
# the line in the middle represents mean the have body or distace form the mean line to the end of box is the standard deviation 
# majority of your data are in the box it self and minorty on the tails and outliers are the points 
```

```python
# to connect tails up to the outliers use whis argumnes as follows
sns.boxplot(data = [arr1 ,  arr2] , whis = np.inf)
```

```python
# of course using only the box plot can show you if your data is normalized or not 

# lets normalize our consider two arrays with normalized data as follows
from scipy import stats 
```

```python
data1 = stats.norm(0,5).rvs(100)

data2 = np.concatenate([stats.gamma(5).rvs(50) - 1 , -1 * stats.gamma(5).rvs(50)])


# dont worry about how I create the data for ploting its just to show you standarized , normalized feature would look like in box plot
# data1 is not normally distributed and data2 is
# look how the tails will be different 
# how mean is exactly at zero on data2 box plot 
# how data and standard deviation is equal on both sides for data2 and not the same for data1
#if you dont see what am talking about you can run this cell serveral times
sns.boxplot(data = [data1 , data2] , whis = np.inf)
```

```python
# one thing that you may have noticed till now .. where is the  KDE to identify really the shape of ditripution
# Here we go with violen
# the shap is so alike violen but its a box plot with cureved border of KDE curve shape
sns.violinplot(data = [data1 , data2])
```

# 5. heat maps

```python
# heat maps is visualization that really helps u identify values that relay featrues with each other
# many data scientists beleive that is fastest way and the short cut for identifying correlation between features 
# heat maps can realy help you selecting your models and desciding weather your values needs to be normalized or not

# in this section we are going to use one of the builtin datasets that comes aside with seaborn

flight_df = sns.load_dataset('flights')
flight_df
```

```python
# lets pivot data to understand it better
pivot_df =   pd.pivot_table(flight_df, index = 'month' , columns = 'year' , values = 'passengers' , aggfunc='sum')
pivot_df
```

```python
# mmm dont you think that this dataset is hard to look at the bigger values 
# what if we can assing a color batch for each values
sns.heatmap(pivot_df)
```

```python
# now whith one look you can see that jul and aug 1960 is the biggerst values amoung all pivoted data
# for futher help you can show annotaion on each batch
sns.set(rc={'figure.figsize':(25.7,8.27)}) # if you want to chnage figure size
sns.heatmap(pivot_df , annot= True , fmt='g') # try to run this line without the parmeter fmt 
```
